```python
# reqeusts, bs4 import
import requests 
import bs4
# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup
```

## 1. Daum 뉴스기사 제목 스크래핑하기
### 1) 질문:  아래의 url에서 뉴스기사의 링크와 제목을 출력하세요.

```python
url = 'https://news.daum.net/economy'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
}

res = requests.get(url, headers=headers)
print(url)
print(type(res))
print(res.status_code)

if res.ok:
    # 인코딩
    res.encoding = 'utf-8'
    
    soup = BeautifulSoup(res.text, 'html.parser')

    news_items = soup.select('a.item_newsheadline2')
    print(type(news_items), len(news_items))

    for item in news_items:
        link = item['href']
        title_tag = item.select_one('strong.tit_txt')
        title = title_tag.text.strip() 

        print(link)
        print(title)
else:
    print(f'error code: {res.status_code}')
```

### 2) 질문:  여러개의 section 중 하나를 선택해서 url에서 뉴스기사의 링크와 제목을 출력하는 코드를 함수로 작성하기


```python
section_dict = {
    '기후/환경': 'climate',
    '사회': 'society',
    '경제': 'economy',
    '정치': 'politics',
    '국제': 'world',
    '문화': 'culture',
    '생활': 'life',
    'IT/과학': 'tech',
    '인물': 'people'
}

def print_news(section_name):
    section_eng = section_dict.get(section_name)

    url = f'https://news.daum.net/{section_eng}'
    print(f"======> {url} {section_name} 뉴스 <======")

    headers = {
        'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/104.0.0.0 Safari/537.36'
    }

    res = requests.get(url, headers=headers)

    if res.ok:
        res.encoding = 'utf-8'
        soup = BeautifulSoup(res.text, 'html.parser')

        news_items = soup.select('a.item_newsheadline2')

        if not news_items:
            print("[정보] 뉴스 기사를 찾을 수 없습니다.")
            return

        for item in news_items:
            link = item['href']
            title_tag = item.select_one('strong.tit_txt')
            title = title_tag.text.strip() 
            print(link)
            print(title)
    else:
        print(f"에러 코드: {res.status_code}")

print_news('경제')
print_news('인물')
```
